```python
import re
NON_PERSIAN_RE = re.compile(r'[^\u0600-\u06FF\s\d\.,،؟!]')
EXTRA_SPACES_RE = re.compile(r'(?<!\n)\s{2,}(?!\n)')
DIACRITICS_RE = re.compile(r'[\u064B-\u065F\u0670\u06D6-\u06DC\u06DF-\u06E8\u06EA-\u06ED]')
TATWEEL_RE = re.compile(r'ـ+')
EXTRA_NEWLINES_RE = re.compile(r'\n{2,}')

TRANSLATION_TABLE = str.maketrans({
    'ي': 'ی', 'ك': 'ک',
    '٠': '0', '١': '1', '٢': '2', '٣': '3', '٤': '4',
    '٥': '5', '٦': '6', '٧': '7', '٨': '8', '٩': '9'
})
```

```python
def clean_text(text):
    """Clean and normalize Persian poetry text"""
    if not isinstance(text, str) or not text.strip():
        return ""

    text = text.strip()
    text = TATWEEL_RE.sub('', text)
    text = DIACRITICS_RE.sub('', text)
    text = NON_PERSIAN_RE.sub('', text)
    text = EXTRA_SPACES_RE.sub(' ', text)
    # text = EXTRA_NEWLINES_RE.sub('\n', text) # change later
    text = text.translate(TRANSLATION_TABLE)

    return text.strip()
```

```python
df_masnaviat['text_cleaned'] = df_masnaviat["text"].apply(clean_text)
df_ghazaliat['text_cleaned'] = df_ghazaliat["text"].apply(clean_text)
```

```python
SPECIAL_TOKENS = ['<PAD>', '<UNK>', '<START>', '<END>', '<LINE_BREAK>']
```

```python
def format_robaee(text):
    hemistichs = text.strip().split('\n')
    if len(hemistichs) != 4:
        return None
    return f"<START> {hemistichs[0]} \n {hemistichs[1]} <LINE_BREAK> {hemistichs[2]} \n {hemistichs[3]} <END>"

def format_ghazal(text):
    hemistichs = text.strip().split('\n')
    lines = []
    for i in range(0, len(hemistichs) - 1, 2):
        lines.append(f"{hemistichs[i]} \n {hemistichs[i+1]}")
    if not lines:
        return None
    return "<START> " + " <LINE_BREAK> ".join(lines) + " <END>"

def format_masnavi(text):
    return format_ghazal(text)
```

```python
df_robaee["text_formatted"] = df_robaee["text_cleaned"].apply(format_robaee)
df_robaee = df_robaee.dropna(subset=["text_formatted"])

df_ghazaliat["text_formatted"] = df_ghazaliat["text_cleaned"].apply(format_ghazal)
df_ghazaliat = df_ghazaliat.dropna(subset=["text_formatted"])

df_masnaviat["text_formatted"] = df_masnaviat["text_cleaned"].apply(format_masnavi)
df_masnaviat = df_masnaviat.dropna(subset=["text_formatted"])
```

```python
df_robaee["text_formatted"] = "<STYLE:ROBAEE> " + df_robaee["text_formatted"]
df_ghazaliat["text_formatted"] = "<STYLE:GHAZAL> " + df_ghazaliat["text_formatted"]
df_masnaviat["text_formatted"] = "<STYLE:MASNAVI> " + df_masnaviat["text_formatted"]
```

```python
def check_dots(text):
    ''' check if dot's appeared in df '''
    if not isinstance(text, str):
        return False


    pattern = r'\.{2,}' # Pattern to match 2 or more consecutive dots
    matches = re.findall(pattern, text)
    return len(matches) > 0

df_all['has_consecutive_dots'] = df_all['text_cleaned'].apply(check_dots)

dot_count = df_all['has_consecutive_dots'].sum()
print(f"Number of entries with consecutive dots (.., ..., etc.): {dot_count}")

examples = df_all[df_all['has_consecutive_dots']].head(5)
print("\nExamples of text with consecutive dots:")
for i, row in examples.iterrows():
    print(f"Row {i}: {row['text_cleaned']}")

def count_specific_dots(text):
    if not isinstance(text, str):
        return {}

    results = {}
    for i in range(2, 6):  # Check for 2, 3, 4, 5 dots
        pattern = r'\.{' + str(i) + '}'
        matches = re.findall(pattern, text)
        results[f'{i}_dots'] = len(matches)

    # For 6 or more dots
    pattern = r'\.{6,}'
    matches = re.findall(pattern, text)
    results['6+_dots'] = len(matches)

    return results

dot_patterns = df_all['text_cleaned'].apply(count_specific_dots).apply(pd.Series).fillna(0)
pattern_sums = dot_patterns.sum()
print("\nDistribution of consecutive dot patterns:")
for pattern, count in pattern_sums.items():
    print(f"{pattern}: {int(count)}")
```

```python

def has_consecutive_dots(text):
    if not isinstance(text, str):
        return False

    # Pattern to match 2 or more consecutive dots
    pattern = r'\.{2,}'
    return bool(re.search(pattern, text))

total_before = len(df_all) # Get count before filtering

df_all_clean = df_all[~df_all['text_cleaned'].apply(has_consecutive_dots)]

total_after = len(df_all_clean)
removed_count = total_before - total_after

print(f"Records before: {total_before}")
print(f"Records after: {total_after}")
print(f"Records removed: {removed_count}")
```
